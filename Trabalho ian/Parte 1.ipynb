Parte 1:
Nesta parte do trabalho vamos calcular o valor da área entre f(x) e g(x).
Começamos por utilizar a sugestão proposta no enunciado e fazemos h(x) = f(x)-g(x) e igualamos h(x) a zero.
Para calcular os extremos de integração usamos o função fsolve da biblioteca scipy mais expecificamente a spicy.otimize.
Para calcularo Valor do integral usamos a regra de Simpson.

Exemplo 1:
Neste exemplo as funções são f(x) = exp(x) - 1 e g(x) = 5 - X^2.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
def f(x):
    return np.exp(x) - 1
def g(x):
    return 5 - x**2
def h(x) :
    return f(x)-g(x)

initial_guesses = [-2, 2]
zeros = fsolve(h, initial_guesses)

print("Zeros de h(x):\n", zeros)
a = zeros[0]
b = zeros[1]
x_values = np.linspace(-3, 3, 400)
f_values = f(x_values)
g_values = g(x_values)
h_values = h(x_values)
plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label='$f(x) = e^x - 1$', color='blue')
plt.plot(x_values, g_values, label='$g(x) = 5 - x^2$', color='green')
plt.plot(x_values, h_values, label='$h(x) = f(x) - g(x)$', color='red')
plt.axhline(0, color='black', label='$y=0$')
plt.axvline(0, color='black', label='$x=0$')
```

```python
def RegraSimpson(m,a,b,n):
    h = (b -a) / n
    integral = m(a) + m(b) 

    for i in range(1, n):
        if i% 2 == 0:
            integral += 2 * m(a + i *h)
        else:
            integral += 4 * m(a + i *h)
    resultado = (h/3)*integral
    return resultado
print('Valor do integral:\n',RegraSimpson(h, a, b, 100))

def Regra_Simpson_erro(m, a, b, n, majorante):
    h = (b-a) / n
    erro = (h**4 *(b-a)/180) * majorante
    soma = 0
    for i in range (1, n):
        if i% 2 == 0:
            soma = soma + 2*m(a + i*h)
        else:
            soma = soma + 4*m(a + i *h)
    soma = soma + m(a) + m(b)
    return erro
print('Erro de Simpson: \n',Regra_Simpson_erro(h, a, b, 100, 4.04021815))
```

O valor do integral é na realidade o simetrico do valor dado pois area são sempre maiores ou iguais a zero.
Para encontrar o valor do majorante da quarta derivada de h(x) utilizamos o Geogebra.
O erro obtido é muito próximo de zero logo é um erro aceitável.
